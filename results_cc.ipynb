```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.colors import CenteredNorm
%matplotlib inline
from sklearn import metrics
from scipy import stats
```

```python
dfall = pd.read_csv('cc_network_data.csv')
dfall['H'] = -np.log(dfall['bias'])*dfall['bias'] - np.log(1-dfall['bias'])*(1-dfall['bias'])
dfall['HKe'] = dfall['H']*dfall['ke']
dfall['variance']=(1-dfall['bias'])*dfall['bias']

dfall['HU'] = -np.log(dfall['biasUnweighted'])*dfall['biasUnweighted'] - np.log(1-dfall['biasUnweighted'])*(1-dfall['biasUnweighted'])
dfall['HKeU'] = dfall['HU']*dfall['ke']
dfall['varianceU']=(1-dfall['biasUnweighted'])*dfall['biasUnweighted']
dfall['dcbool']=(dfall['dc']>1)+-1*(dfall['dc']<1)
```

```python
fig, ax = plt.subplots(figsize=(10, 10),facecolor='white')
ax.set_facecolor('white')
varx = 'bias'
vary = 'k'
cmap = plt.cm.get_cmap('coolwarm')#.reversed()
sc=ax.scatter(dfall[varx],dfall[vary],s=100,c=dfall['dc'],cmap=cmap,
              norm=CenteredNorm(vcenter=1,halfrange=0.5),alpha=1.0)
px = np.arange(0,1,0.01)
py=1/(2*px*(1-px))
ax.plot(px,py)
ax.set_ylim(0,8)
plt.colorbar(sc)
plt.xlabel(varx)
plt.ylabel(vary)
plt.title('Derrida Coefficient')
plt.show()
```
